```python
from config import key, g_key
from citipy import citipy
import numpy as np
import pandas as pd
import requests
```

```python
# Generate 2000 random lat and lng
lat = np.random.uniform(-90,90,size=2000)
lng = np.random.uniform(-90,90,size=2000)
lat_lng = zip(lat,lng)
coords = list(lat_lng)

# get nearest city name
cities = list()
for coord in coords:
    cities_name = citipy.nearest_city(coord[0],coord[1]).city_name
    if cities_name not in cities:
        cities.append(cities_name)
```

```python
# Perform an API call with the OpenWeatherMap.
record_count = 1
set_count = 1
city_data = list()

basic_url = 'http://api.openweathermap.org/data/2.5/weather?units=imperial&APPID=' + key

for i, item in enumerate(cities):
    url = basic_url + '&q=' + item
    print(f'Processing Record{item}')
    try:
        js = requests.get(url).json()
        city_name = js['name']
        city_country = js['sys']['country']
        city_lat = js['coord']['lat']
        city_lon = js['coord']['lon']
        city_max_temp = js['main']['temp_max']
        city_humidity = js['main']['humidity']
        city_cload = js['clouds']['all']
        city_wind_speed = js['wind']['speed']
        city_weather_description = js['weather'][0]['description']
        city_data.append(
            {'City':city_name, 
             'Country':city_country, 
             'Lat':city_lat, 
             'Lng': city_lon,
             'Max Temp':city_max_temp,
             'Humidity':city_humidity,
             'Cloudiness': city_cload,
             'Wind Speed':city_wind_speed,
             'Current Description':city_weather_description})
    except:
        print("City not found")
        pass
print('------Done-----')
```

```python
city_df = pd.DataFrame(city_data)
city_df.head()
```

```python
city_df.to_csv("./WeatherPy_Database.csv",index=False)
```
